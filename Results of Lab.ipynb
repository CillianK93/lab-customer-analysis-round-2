```python
import pandas as pd
import numpy as np
df_1=pd.read_csv('marketing_customer_analysis.csv')
df_1
```

## Dataframe Shape

```python
df_1.shape
```

## Standardizing Columns

```python
df_1.columns = [col_name.lower().replace(" ", "_") for col_name in df_1.columns]
df_1.columns
```

```python
df_1= df_1.drop(columns=['unnamed:_0'])
df_1=df_1.rename(columns={'employmentstatus':'employment_status'})
df_2=df_1.copy
df_1

```

## Numerical and Categorical columns

```python
df_1.dtypes
```

## Checking and dealing with NaN values

```python
df_1.isna().sum()
```

## State NaN

```python
df_1[df_1['state'].isna()]
```

```python
df_1['state'] = df_1['state'].fillna('')
```

## Response NaN

```python
df_1[df_1['response'].isna()]
```

```python
df_1['response']=df_1['response'].fillna('No')
```

## Months since last claim NaN

```python
df_1[df_1['months_since_last_claim'].isna()]
```

```python
col_median =df_1['months_since_last_claim'].median()
df_1['months_since_last_claim']=df_1['months_since_last_claim'].fillna(col_median)
```

## Number of open complaints NaN

```python
df_1[df_1['number_of_open_complaints'].isna()]
```

```python
df_1['number_of_open_complaints'] =df_1['number_of_open_complaints'].fillna('0')
```

## Vehicle class and size NaNs

```python
df_1['vehicle_class'] =df_1['vehicle_class'].fillna('')
df_1['vehicle_size'] =df_1['vehicle_size'].fillna('')
```

## Vehicle type NaN

```python
#because the NaNs are so high I will drop the column
df_1=df_1.drop(columns=['vehicle_type'])
```

```python
df_1.isna().sum()
```

## Datetime Format

```python
df_1['effective_to_date']=pd.to_datetime(df_1['effective_to_date'],errors='coerce')
df_1['effective_to_date']
```

```python
df_1['month'] =df_1['effective_to_date'].dt.month
```

```python
df_1[df_1['month']<2]
```
